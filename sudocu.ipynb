```python
import sys
import math
import gurobipy as gp
from gurobipy import GRB
```

```python
if len(sys.argv) < 2:
    print('Usage: sudoku.py filename')
    sys.exit(0)

fileNames = ["sudoku data-20240313T145207Z-001/sudoku data/sudoku1",
             "sudoku data-20240313T145207Z-001/sudoku data/sudoku2",
             "sudoku data-20240313T145207Z-001/sudoku data/sudoku3",
             "sudoku data-20240313T145207Z-001/sudoku data/sudoku4",
             "sudoku data-20240313T145207Z-001/sudoku data/sudoku5",
             "sudoku data-20240313T145207Z-001/sudoku data/sudokuGen1",
             "sudoku data-20240313T145207Z-001/sudoku data/sudokuGen2"]
f = open(fileNames[0])

grid = f.read().split()

n = len(grid[0])
s = int(math.sqrt(n))
```

```python
model = gp.Model('sudoku')

vars = model.addVars(n, n, n, vtype=GRB.BINARY, name='G')
```

```python
for i in range(n):
    for j in range(n):
        if grid[i][j] != '.':
            v = int(grid[i][j]) - 1
            vars[i, j, v].LB = 1
```

```python
model.addConstrs((vars.sum(i, j, '*') == 1
                 for i in range(n)
                 for j in range(n)), name='V')
```

```python
model.addConstrs((vars.sum(i, '*', v) == 1
                 for i in range(n)
                 for v in range(n)), name='R')
```

```python
model.addConstrs((vars.sum('*', j, v) == 1
                 for j in range(n)
                 for v in range(n)), name='C')
```

```python
model.addConstrs((
    gp.quicksum(vars[i, j, v] for i in range(i0*s, (i0+1)*s)
                for j in range(j0*s, (j0+1)*s)) == 1
    for v in range(n)
    for i0 in range(s)
    for j0 in range(s)), name='Sub')
```

```python
times = []

for j in range (6):
    model.Params.Method = j
    time = []
    for i in range(100):
        model.optimize()
        time.append(model.Runtime)
        model.reset()
    times.append(time)
    
```

```python
import numpy as np
dict_methods = {0: "primal simplex", 1: "dual simplex", 2: "barrier",3: "concurrent", 4: "deterministic concurrent", 5: "deterministic concurrent simplex"}
for i in range(6):
    print(f'for method \"{dict_methods.get(i)}\":\nmean value: {np.mean(np.array(times[i]))}\nvarience: {np.var(np.array(times[i]))}')
```
